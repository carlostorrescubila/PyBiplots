# Packages

```python
import pybiplots.HJ_Biplot as hj
import statsmodels.api as sm
import matplotlib
```

# Data

```python
mtcars = sm.datasets.get_rdataset('mtcars').data
mtcars.head()
```

# Fit biplot method

In this example is used HJ-Biplot.

```python
HJ = hj.fit(mtcars, Transform='Standardize columns')
```

# Biplot

We set the graphic style to seaborn style. To see more styles visit https://matplotlib.org/3.2.2/gallery/style_sheets/style_sheets_reference.html

```python
matplotlib.style.use("seaborn")
```

## 1. Default

```python
HJ.plot()
```

## 2. Adjust text

```python
HJ.plot(adjust_ind_name=True)
```

## 3. Font size

```python
HJ.plot(adjust_ind_name=True, name_fontsize=10, axis_fontsize=25)
```

## 4. Add or delete points and vector labels

```python
HJ.plot(ind_name=False, vec_name=True)
```

## 5. Select axis

```python
HJ.plot(ind_name=False, axis=(1,3))
```

## 6. Show groups

```python
HJ.plot(ind_name=False, groups=mtcars['vs'], convex_hull=True)
```

## 7. Select colors

```python
HJ.plot(adjust_ind_name=True, palette=['green'])
```

```python
HJ.plot(
    adjust_ind_name=True, 
    groups=mtcars['vs'], 
    convex_hull=True,
    palette=['green', 'blue'],
    vec_color='red'
)
```
