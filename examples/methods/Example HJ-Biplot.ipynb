# HJ-Biplot

## 1. Packages

```python
import pybiplots.HJ_Biplot as hj
import pandas as pd
import matplotlib
```

## 2. Load data

```python
mtcars = pd.read_csv('mtcars.txt', delimiter = "\t")
mtcars = mtcars.set_index("Unnamed: 0")
mtcars.index.name = None
```

## 3. Fit HJ-Biplot

```python
HJ = hj.fit(mtcars, Transform = 'Standardize columns')
```

## 4. Results

### Eigenvalues

```python
HJ.eigenvalues.round(2)
```

### Explained variance

```python
HJ.explained_variance.round(2)
```

### Loadings

```python
HJ.loadings.round(2)
```

### Row coordinates

```python
HJ.row_coordinates.round(2).head()
```

### Column coordinates

```python
HJ.column_coordinates.round(2)
```

### Row contributions

```python
HJ.row_contributions.head() 
```

### Column contributions

```python
HJ.column_contributions
```

### Biplot

```python
matplotlib.style.use('seaborn')
HJ.plot(adjust_ind_name = True)
```
