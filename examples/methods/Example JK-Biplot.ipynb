# JK-Biplot

## 1. Packages

```python
import pybiplots.JK_Biplot as jk
import pandas as pd
import matplotlib
```

## 2. Load data

```python
mtcars = pd.read_csv('mtcars.txt', delimiter="\t")
mtcars = mtcars.set_index("Unnamed: 0")
mtcars.index.name = None
```

## 3. Fit JK-Biplot

```python
JK = jk.fit(mtcars, Transform = 'Standardize columns')
```

## 4. Results

### Eigenvalues

```python
JK.eigenvalues.round(2)
```

### Explained variance

```python
JK.explained_variance.round(2)
```

### Loadings

```python
JK.loadings.round(2)
```

### Row coordinates

```python
JK.row_coordinates.round(2).head()
```

### Column coordinates

```python
JK.column_coordinates.round(2)
```

### Row contributions

```python
JK.row_contributions.head() 
```

### Column contributions

```python
JK.column_contributions
```

### Biplot

```python
matplotlib.style.use("seaborn")
JK.plot(adjust_ind_name = True, name_fontsize=13)
```
