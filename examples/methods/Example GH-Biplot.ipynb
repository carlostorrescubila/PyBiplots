# GH-Biplot

## 1. Packages

```python
import pybiplots.GH_Biplot as gh
import pandas as pd
import matplotlib
```

## 2. Load Data

```python
mtcars = pd.read_csv('mtcars.txt', delimiter="\t")
mtcars = mtcars.set_index("Unnamed: 0")
mtcars.index.name = None
```

## 3. Fit GH-Biplot

```python
GH = gh.fit(mtcars, Transform = 'Standardize columns')
```

## 4. Results

### Eigenvalues

```python
GH.eigenvalues.round(2)
```

### Explained variance

```python
GH.explained_variance.round(2)
```

### Loadings

```python
GH.loadings.round(2)
```

#### Row coordinates

```python
GH.row_coordinates.round(2).head()
```

#### Column coordinates

```python
GH.column_coordinates.round(2)
```

### Row contributions

```python
GH.row_contributions.head()
```

### Column contributions

```python
GH.column_contributions
```

### Biplot

```python
matplotlib.style.use("seaborn")
GH.plot(adjust_ind_name = True, name_fontsize=13)
```
